##Нахождение похожих предложений

```
import numpy as np
from scipy.spatial import distance
import re
```

```
with open("/content/drive/My Drive/Colab Notebooks/MPTI_ML_1course/sentences.txt") as f:
  # определяем количество строк, предложений в текстовом файле
  lines = sum(1 for _ in f)
  f.seek(0)
        
  words = {}
  # переменная 'lcount' контролирует индексы текущей строки в текстовом файле
  # переменная 'wcount' контролирует текущее уникальное слово.
  lcount, wcount = 0, 0
  for line in f:
    # записываем переменную строки, переводим все в нижний регистр
    p = line.lower()
    # проводим токенизацию, т.е. разделение предложения на слова
    tokens = re.split('[^a-z]', p)
    # пробегаем по всему списку, оставляя только cлова, иначе просто выкидываем пробелы
    tokens = [x for x in tokens if x]
    for token in tokens:
      # Если слово еще не представлено в словаре, то мы его добавляем.
      # ключ "occurrences" - это объект списка, где каждый элемент представляет собой
      # определенную строку в текстовом файле. Первый элемент представляет собой первую строку, второй элемент
      # представляет собой вторую строку и т. д.
      if token not in words:
        words[token] = {
            "index": wcount,
            "occurrences": [0] * lines
            }
        wcount += 1
        # Если мы находим одно и то же слово в одном и том же списке 
        # слов (в одной и той же строке), то мы просто игнорируем его.
      elif words[token]["occurrences"][lcount] != 0:
        continue
        # здесь мы записываем сколько раз это слово встречается в каждом предложении (строке)  
        # count() -- выдает количество совпадений слова в данном предложении. 
      words[token]["occurrences"][lcount] = tokens.count(token)    
      # print(words[token]["occurrences"][lcount],'\n',tokens.count(token))
    lcount += 1

  wordM = np.zeros((lines, len(words)))
  # Теперь для каждого слова в словаре мы берем число его 
  # вхождений в каждой строке и помещаем его в массив. 
  for word in words:
    i, j = 0, words[word]["index"]
    for occ in words[word]["occurrences"]:
      wordM[i, j] = occ
      i += 1
      
  # расчитываем косинусное расстояние между первым предложением 
  # и другими предложениями в тексте
  dist = [] 
  u = wordM[0,:] 
  for i in range(1, lines):
    v = wordM[i,:]
    dist.append({"index": i, "distance": distance.cosine(u, v)}) 
  
  dist.sort(key=lambda x: x["distance"])
  print("The 1st closest sentence is a sentence #%d with a cosine distance of %.2f.\n"\
        "The 2nd closest sentence is a sentence #%d with a cosine distance of %.2f." % (
        dist[0]["index"],
        dist[0]["distance"],
        dist[1]["index"],
        dist[1]["distance"]
        ))
```

##Аппроксимация функции

```
import numpy as np
import math
import scipy as sp
```

```
def f(x):
    return (math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2))

x = np.array([1, 4, 10, 15])
b = np.array([f(a) for a in x])[:,np.newaxis]

A = np.zeros((4,4))
for j in range(len(x)):
  A[j,] = 1, x[j], x[j]**2, x[j]**3

w = sp.linalg.solve(A,b)
print(np.around(w, 2))
```
